```python
X = [2015 , 2016 , 2017 , 2018 , 2019 , 2020]
Y = [15 , 16 , 16.5 , 18 , 19.5 , 22]
plt.plot(X , Y) # plot() x-axis , y-axis   value = list/array/np.array
plt.show() # Display 
```

```python
Prices = [10 , 15 , 20 ,35 , 36, 25, 45,50 , 60 , 80 ,85,95, 100 ,120 ,130,140]
plt.hist(Prices , bins = 7) # y axis , bins 
plt.show()
```

```python
# max = 140 
# bins = 7
# max / bins   140 / 7 = 20
# min-value  10-30 30-50 50-70 70-90 90-110 110-130 130-140
```

```python
X = [2015 , 2016 , 2017 , 2018 , 2019 , 2020]
Y = [15 , 16 , 16.5 , 18 , 19.5 , 22]
plt.plot(X , Y , color = "red") # plot() x-axis , y-axis   value = list/array/np.array
plt.scatter(X , Y , color = "violet")
plt.xlabel("Year" , color = "Purple")
plt.ylabel("Poppulation Per Year" , color = "blue")
plt.title("Relationship Of Year And Poppulation" , color = "orange")
plt.yticks([15 , 15.5 , 16 , 16.5 , 17 , 17.5 , 18 , 18.5 , 19 , 19.5 , 20 , 20.5 , 21 , 21.5 , 22 , 22.5 , 23] , color = "green")
plt.xticks([2015 , 2017 , 2019 , 2021] ,color = "Brown")
plt.show() # Display
```

```python
X = [2015 , 2016 , 2017 , 2018 , 2019 , 2020]
Y = [15 , 16 , 16.5 , 18 , 19.5 , 22]
plt.plot(X , Y , color = "red" , linewidth = 3 , linestyle = ":" , marker = "*") # plot() x-axis , y-axis   value = list/array/np.array
#plt.scatter(X , Y , color = "violet")
plt.xlabel("Year" , color = "Purple")
plt.ylabel("Poppulation Per Year" , color = "blue")
plt.title("Relationship Of Year And Poppulation" , color = "orange")
plt.yticks([15 , 15.5 , 16 , 16.5 , 17 , 17.5 , 18 , 18.5 , 19 , 19.5 , 20 , 20.5 , 21 , 21.5 , 22 , 22.5 , 23] , color = "green")
plt.xticks([2015 , 2017 , 2019 , 2021] ,color = "Brown")
plt.show() # Display
```
