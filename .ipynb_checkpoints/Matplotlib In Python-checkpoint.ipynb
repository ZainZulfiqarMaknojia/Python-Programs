```python
# Matplotlib 
import matplotlib.pyplot as plt
Month = ["Dec" , "Jan" , "Feb" ,"March" , "April" , "May" , "Jun" , "July" , "August" , "September"]
Active = [0 , 0 , 2 , 100 , 500 , 1400 , 2300 , 5000 , 4900 , 3000]
plt.plot(Month , Active)
plt.show()
```

```python
year = [2011, 2012, 2013,2014,2015,2016,2017,2018,2019,2020]
Profit = [1000,2000,5000,10000,60000,30000,31000,50000,55000,70000]
plt.plot(year,Profit)
plt.show()
```

```python
year = [2011, 2012, 2013,2014,2015,2016,2017,2018,2019,2020]
Profit = [1000,2000,5000,10000,60000,30000,31000,50000,55000,70000]
plt.plot(year,Profit)
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```

```python
year = [2011, 2012, 2013,2014,2015,2016,2017,2018,2019,2020]
Profit = [1000,2000,5000,10000,60000,30000,31000,50000,55000,70000]
plt.plot(year,Profit)
plt.xlabel("Year")
plt.ylabel("Profit")
plt.title("Amazon Net Profit Per Year")
plt.show()
```
